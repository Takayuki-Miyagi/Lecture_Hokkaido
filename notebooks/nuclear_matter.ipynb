## 核物質
ここでは核物質について考えてみましょう．
核物質とは，無限に広がった密度一定の核子系で，核子間相互作用のみで記述される理想化された系で，いわば無限に大きい原子核です．
無限に大きい原子核なんて考えてもしょうがないじゃないかと思うかもしれませんが，実は中性子星は限りなく核物質に近い状態であると考えられています．
また，理想化された系を考えることで，有限核の性質を理解するための重要な手がかりを得ることもできます．

核物質について考える時に中心的な量はやはりエネルギーです．
無限系なので，核子あたりのエネルギー$\varepsilon \equiv E/A$を考えるのが自然です．
また系全体で密度は一定なので，それをパラメータとして$\varepsilon(\rho)$という関数を考えます．
さらに一般化して，中性子と陽子の比率をパラメータとして加えることもできますが，ここでは単に，全てが中性子の中性子物質$\varepsilon_{n}(\rho)$と中性子と陽子が同数の対称核物質$\varepsilon(\rho)$を考えます．
対称核物質について，すでに有限核で得られている知見を利用することができます．
原子核には飽和性，つまり原子核の中心密度と核子あたりのエネルギーがほぼ一定であるという性質があります．
このことから，有有限の情報を通じて我々がアクセスできるのは飽和密度$\rho_{0}$の近辺ということになります．
また，どの原子核でも中心密度が一定ということは，密度を変化させるとエネルギー的に損をするということです．
つまり，
\begin{equation*}
\frac{d\varepsilon}{d\rho} \bigg|_{\rho = \rho_{0}} = 0
\end{equation*}
です．また，有限核の実験から，原子核の密度を変えることは非常に難しいこともわかっています．つまり，
\begin{equation*}
\frac{d^{2}\varepsilon}{d\rho^{2}} \bigg|_{\rho = \rho_{0}} 
\end{equation*}
は比較的大きい値を取ることがわかります．
つまり，実質的には
\begin{equation*}
\varepsilon(\rho) \approx \varepsilon(\rho_{0}) + \frac{K}{2} \left(\frac{\rho - \rho_{0}}{3\rho_{0}}\right)^{2}
\end{equation*}
くらいで展開しても良さそうです．
これに対応して，中性子核物質についても
\begin{equation*}
\varepsilon_{n}(\rho) \approx \varepsilon_{n}(\rho_{0}) + L_{n} \left(\frac{\rho - \rho_{0}}{3\rho_{0}}\right) + \frac{K_{n}}{2} \left(\frac{\rho - \rho_{0}}{3\rho_{0}}\right)^{2}
\end{equation*}
と展開されます．
こんなことして何が嬉しいのでしょうか？
実は中性子星の物理と関係しています．

Einstein方程式は
\begin{equation*}
G_{\mu\nu} = 8\pi G T_{\mu\nu}
\end{equation*}
として与えられます．ここで，$G_{\mu\nu}$はEinsteinテンソルで，$T_{\mu\nu}$はエネルギー運動量テンソルです．
球対称な計量を使うと動径方向の圧力勾配の方程式として
\begin{equation*}
\frac{dP}{dr} = - \frac{G M(r) \rho(r)}{r^{2}} \left(1 + \frac{P(r)}{\rho(r)}\right) \left(1 + \frac{4\pi r^{3} P(r)}{M(r)}\right) \left(1 - \frac{2GM(r)}{r}\right)^{-1}
\end{equation*}
が得られます．この方程式はTolman-Oppenheimer-Volkoff (TOV)方程式として知られています．
この式が何を意味にしているのか，見るために$c$を復活させましょう．
すると
\begin{equation*}
\frac{dP}{dr} = - \frac{G M(r) \rho(r)}{r^{2}} \left(1 + \frac{P(r)}{\rho(r)c^{2}}\right) \left(1 + \frac{4\pi r^{3} P(r)}{M(r)c^{2}}\right) \left(1 - \frac{2GM(r)}{rc^{2}}\right)^{-1}
\end{equation*}
直感的にわかりやすくするために，非相対論的な極限$c \to \infty$を考えると，$1/c^{2}$に比例する項は相対論的な補正項とみなして無視できます．すると
\begin{equation*}
\frac{dP}{dr} = - \frac{G M(r) \rho(r)}{r^{2}}
\end{equation*}
という非常に単純な式が得られます．この式は半径$r$の球の中に入っている質量$M(r)$によるニュートン重力がと圧力勾配の釣り合いを表しています．
ここで，内部の質量$M(r)$は
\begin{equation*}
M(r) = \int_{0}^{r} dr' 4\pi r'^{2} \rho(r')
\end{equation*}
として与えられます．
圧力と密度の関係はその系を構成する物質の性質に依存し，圧力と密度の関係を与える状態方程式がわかればTOV方程式と合わせて解くことができます．
そして，圧力は内部エネルギーと関係していて
\begin{equation*}
P = \rho^{2}\frac{d\varepsilon}{d\rho}
\end{equation*}
で与えられます．

ここで，一番簡単な状態方程式について考えてみましょう．
まず，中性子星は中性子のみからできている温度$T=0$のとして，中性子間の相互作用を無視したfree Fermi gasを考えてみましょう．
簡単のために，周期境界条件を課した箱の中で考えます．箱の長さを$L$とすると，波動関数は以下のようになります．
\begin{equation*}
\phi(x,y,z) = \phi_{x}(x) \phi_{y}(y) \phi_{z}(z), \quad \phi_{i}(x_{i}) = \sqrt{\frac{1}{L}} e^{i k_{i} x_{i}}, \quad k_{i} = \frac{2\pi n_{i}}{L}, \quad n_{i} \in \mathbb{Z}
\end{equation*}
十分$L$が大きい極限では，$k_{i}$は連続的に取れます．
箱の中に$N$個の中性子が入っているとすると，$N$個の中性子はFermi球を形成します．
\begin{equation*}
N = 2 \frac{4\pi}{3} k_{F}^{3} \left(\frac{L}{2\pi}\right)^{3} = 
\frac{1}{3\pi^{2}} k_{F}^{3} V
\end{equation*}
この時，運動量の大きさは数密度$\rho = N/V$で与えられます．
\begin{equation*}
k_{F} = (3\pi^{2}\rho)^{1/3}
\end{equation*}
ここで，$k_{F}$はFermi運動量で，$2$は中性子のスピンの自由度です．
原子核の中心密度はだいたい$\rho_{0} \sim 0.16$ fm$^{-3}$で，中性子の密度はその半分の$0.08$ fm$^{-3}$くらいなので，
$k_{F} \sim 1.33$~fm$^{-1} \sim 260$~MeVとなります．
$\left(L/2\pi\right)^{3}$は箱の体積に対して格子点の数に変換するための因子です．
また，$V = L^{3}$は箱の体積です．
中性子星の中では中性子の運動量はその質量に比べて小さいので，非相対論的な分散関係を使うことができます．
\begin{equation*}
\epsilon = \frac{k^{2}}{2m}
\end{equation*}
すると，
\begin{equation*}
N = \frac{V}{3\pi^{2}} (2m\epsilon_{F})^{3/2}
\end{equation*}
したがって，
\begin{equation*}
\epsilon_{F} = \frac{1}{2m} (3\pi^{2}\rho)^{2/3}
\end{equation*}
Fermiエネルギーはだいたい$\epsilon_{F} \sim 40$~MeVとなります．
状態密度は
\begin{equation*}
D = \frac{dN}{d\epsilon} = \frac{V}{2\pi^{2}} (2m)^{3/2} \epsilon^{1/2}, \quad
\end{equation*}
系の全エネルギーは
\begin{equation*}
E = \int_{0}^{\epsilon_{F}} d\epsilon \, \epsilon D(\epsilon) = \frac{V}{2\pi^{2}} (2m)^{3/2} \frac{2}{5} \epsilon_{F}^{5/2}
\end{equation*}
となるので，核子あたりのエネルギーは
\begin{equation*}
\varepsilon = \frac{3}{5} \epsilon_{F} = \frac{3}{5} \frac{1}{2m} (3\pi^{2}\rho)^{2/3}
\end{equation*}
となります．
このことから，非相対論的なfree Fermi gasの状態方程式は
\begin{equation*}
P = \rho^{2} \frac{d\varepsilon}{d\rho} = \frac{2}{5} \frac{1}{2m} (3\pi^{2})^{2/3} \rho^{5/3}
\end{equation*}
となります．ちなみに，飽和密度$\rho_{0}$での圧力は$P = \frac{\rho_{0}}{3} L_{n}$とparametrizeされるので，
だいたい$L_{n} = \frac{6}{5} \epsilon_{F}  \approx 45$~MeVとなります．

得られた$\varepsilon$をプロットすると以下のようになります．

```python
from matplotlib import pyplot as plt
import numpy as np
m_neutron = 939.565 # MeV
hc = 197.327 # MeV fm
rho = np.linspace(0, 0.3, 100)
plt.plot(rho, (3/5) * (1/(2*m_neutron)) * (3*np.pi**2 * rho)**(2/3) * hc**2)
plt.xlim(0,0.3)
plt.ylim(0,60)
plt.xlabel(r"$\rho$ (fm$^{-3}$)")
plt.ylabel(r"$\varepsilon$ (MeV)")
```

それでは，Fermiガス模型を使っTOV方程式を解いてみましょう．
本当の中性子星には電子や陽子も入っているので，電気的に中性になるようにもう一つ方程式を連立して解く必要がありますが，ここでは簡単のために無視します．
方程式は$r=0$で中心密度与えて外側に向かって$P(r)=0$となる点まで積分していくことで解くことができます．
今はChatGPTとかに式を投げればコードを書いてくれるのでとても簡単です．

```python
from scipy.integrate import solve_ivp
from scipy.constants import G, c, hbar, physical_constants

# ============================================================
# Physical constants
# ============================================================

m = physical_constants["neutron mass"][0]   # neutron mass [kg]
M_sun = 1.98847e30                          # solar mass [kg]

# Nonrelativistic ideal Fermi-gas EOS:
#
#   P = hbar^2 / (5m) * (3 pi^2)^(2/3) * n^(5/3)
#
# With mass density rho = m n:
#
#   P = K rho^(5/3)
#
K = hbar**2 * (3.0 * np.pi**2)**(2.0 / 3.0) / (5.0 * m**(8.0 / 3.0))


def pressure_from_density(rho):
    """
    Pressure as a function of mass density.

    Parameters
    ----------
    rho : float or ndarray
        Mass density [kg/m^3].

    Returns
    -------
    P : float or ndarray
        Pressure [Pa].
    """
    rho = np.asarray(rho)
    return K * rho**(5.0 / 3.0)


def density_from_pressure(P):
    """
    Mass density obtained by inverting P = K rho^(5/3).

    Parameters
    ----------
    P : float or ndarray
        Pressure [Pa].

    Returns
    -------
    rho : float or ndarray
        Mass density [kg/m^3].
    """
    P = np.maximum(P, 0.0)
    return (P / K)**(3.0 / 5.0)


def solve_tov(
    rho_c,
    r_max=1.0e5,
    r0=1.0,
    max_step=100.0,
    rtol=1.0e-8,
    corrections=True,
):
    """
    Solve the TOV equations for a specified central mass density.

    Parameters
    ----------
    rho_c : float
        Central mass density [kg/m^3].
    r_max : float
        Maximum integration radius [m].
    r0 : float
        Initial radius used to avoid the r = 0 singularity [m].
    max_step : float
        Maximum radial step [m].
    rtol : float
        Relative tolerance of the ODE solver.

    Returns
    -------
    result : scipy.integrate.OdeResult
        Numerical solution.
    """

    P_c = pressure_from_density(rho_c)

    # Near the origin:
    #
    # M(r) = 4 pi rho_c r^3 / 3 + O(r^5)
    #
    M0 = (4.0 * np.pi / 3.0) * rho_c * r0**3

    # Leading correction to the pressure near r = 0.
    x_c = P_c / (rho_c * c**2)

    pressure_correction = (
        (2.0 * np.pi * G / 3.0)
        * rho_c**2
        * (1.0 + x_c)
        * (1.0 + 3.0 * x_c)
        * r0**2
    )

    P0 = P_c - pressure_correction

    # y[0] = P(r)
    # y[1] = M(r)
    y0 = np.array([P0, M0])

    def tov_rhs(r, y):
        P, M = y

        # The stellar surface has already been reached.
        if P <= 0.0:
            return np.array([0.0, 0.0])

        rho = density_from_pressure(P)

        compactness_factor = 1.0 - 2.0 * G * M / (r * c**2)

        if compactness_factor <= 0.0:
            raise RuntimeError(
                "1 - 2GM/(rc^2) became non-positive. "
                "The integration approached a horizon."
            )

        pressure_energy_correction = 1.0 + P / (rho * c**2)

        active_mass_correction = (
            1.0 + 4.0 * np.pi * r**3 * P / (M * c**2)
        )

        if(not corrections):
            pressure_energy_correction = 1.0
            active_mass_correction = 1.0
            compactness_factor = 1.0
        dP_dr = (
            -G * M * rho / r**2
            * pressure_energy_correction
            * active_mass_correction
            / compactness_factor
        )

        dM_dr = 4.0 * np.pi * r**2 * rho

        return np.array([dP_dr, dM_dr])

    # Stop when P decreases through zero.
    def surface_event(r, y):
        return y[0]

    surface_event.terminal = True
    surface_event.direction = -1

    solution = solve_ivp(
        fun=tov_rhs,
        t_span=(r0, r_max),
        y0=y0,
        events=surface_event,
        rtol=rtol,
        atol=np.array([1.0e10, 1.0e10]),
        max_step=max_step,
        dense_output=True,
    )

    if not solution.success:
        raise RuntimeError(solution.message)

    if len(solution.t_events[0]) == 0:
        raise RuntimeError(
            "The pressure did not reach zero before r_max. "
            "Increase r_max."
        )

    return solution


# ============================================================
# Example
# ============================================================

rho_c = 8.0e17  # central mass density [kg/m^3]

solution = solve_tov(rho_c)

r = solution.t
P = np.maximum(solution.y[0], 0.0)
M = solution.y[1]
rho = density_from_pressure(P)

R = solution.t_events[0][0]
M_star = solution.y_events[0][0, 1]

compactness = G * M_star / (R * c**2)

print(f"Central density : {rho_c:.6e} kg/m^3")
print(f"Central pressure: {pressure_from_density(rho_c):.6e} Pa")
print(f"Radius          : {R / 1.0e3:.6f} km")
print(f"Mass            : {M_star / M_sun:.6f} M_sun")
print(f"GM/(Rc^2)       : {compactness:.6f}")
print(f"2GM/(Rc^2)      : {2.0 * compactness:.6f}")


# ============================================================
# Plot density, pressure, and enclosed mass
# ============================================================

plt.figure(figsize=(7, 5))
plt.plot(r / 1.0e3, rho)
plt.xlabel(r"$r\ \mathrm{[km]}$")
plt.ylabel(r"$\rho(r)\ \mathrm{[kg/m^3]}$")
plt.yscale("log")
plt.grid()
plt.tight_layout()
plt.show()


plt.figure(figsize=(7, 5))
plt.plot(r / 1.0e3, P)
plt.xlabel(r"$r\ \mathrm{[km]}$")
plt.ylabel(r"$P(r)\ \mathrm{[Pa]}$")
plt.yscale("log")
plt.grid()
plt.tight_layout()
plt.show()


plt.figure(figsize=(7, 5))
plt.plot(r / 1.0e3, M / M_sun)
plt.xlabel(r"$r\ \mathrm{[km]}$")
plt.ylabel(r"$M(r)/M_\odot$")
plt.grid()
plt.tight_layout()
plt.show()

central_densities = np.logspace(
    np.log10(5.0e16),
    np.log10(5.0e19),
    60,
)

radii = []
masses = []
valid_central_densities = []

for rho_c_i in central_densities:
    try:
        sol_i = solve_tov(
            rho_c=rho_c_i,
            r_max=2.0e5,
            max_step=100.0,
            corrections=True,
        )

        R_i = sol_i.t_events[0][0]
        M_i = sol_i.y_events[0][0, 1]

        valid_central_densities.append(rho_c_i)
        radii.append(R_i / 1.0e3)
        masses.append(M_i / M_sun)

    except RuntimeError as error:
        print(f"rho_c = {rho_c_i:.3e}: {error}")
valid_central_densities = np.asarray(valid_central_densities)
central_number_densities_m3 = valid_central_densities / m
central_number_densities_rho0 = central_number_densities_m3 * 1.0e-45 / 0.16
radii = np.asarray(radii)
masses = np.asarray(masses)

radii_N = []
masses_N = []
valid_central_densities_N = []

for rho_c_i in central_densities:
    try:
        sol_i = solve_tov(
            rho_c=rho_c_i,
            r_max=2.0e5,
            max_step=100.0,
            corrections=False,
        )

        R_i = sol_i.t_events[0][0]
        M_i = sol_i.y_events[0][0, 1]

        valid_central_densities_N.append(rho_c_i)
        radii_N.append(R_i / 1.0e3)
        masses_N.append(M_i / M_sun)

    except RuntimeError as error:
        print(f"rho_c = {rho_c_i:.3e}: {error}")

valid_central_densities_N = np.asarray(valid_central_densities_N)
central_number_densities_m3_N = valid_central_densities_N / m
central_number_densities_rho0_N = central_number_densities_m3_N * 1.0e-45 / 0.16
radii_N = np.asarray(radii_N)
masses_N = np.asarray(masses_N)


plt.figure(figsize=(7, 5))
fs = 72 * 0.3
plt.rcParams['font.size'] = fs
plt.rcParams['legend.fontsize'] = fs
plt.rcParams['xtick.labelsize'] = fs
plt.rcParams['ytick.labelsize'] = fs
plt.rcParams['axes.labelsize'] = fs
plt.plot(radii_N, masses_N, marker="o", markersize=3, label="Newton")
plt.plot(radii, masses, marker="o", markersize=3, label="GR")
plt.xlabel(r"$R\ \mathrm{(km)}$")
plt.ylabel(r"$M/M_\odot$")
plt.legend()
plt.tight_layout()
plt.show()


plt.figure(figsize=(7, 5))
plt.plot(central_number_densities_rho0_N, masses_N, marker="o", markersize=3, label="Newton")
plt.plot(central_number_densities_rho0, masses, marker="o", markersize=3, label="GR")
plt.xlabel(r"$\rho_c / \rho_{0}$")
plt.ylabel(r"$M/M_\odot$")
plt.xscale("log")
plt.legend()
plt.tight_layout()
plt.show()
```

答えは，上の図のようになります．
ここで，中心密度が大きいようなところでは，状態方程式を導く時に使った非相対論的な近似が破綻するので，本当は状態方程式には修正が必要なはずですがそれは無視しています．
もし，非相対論的極限を取ったTOV方程式を使った非現実的な計算をすると，中性子星の質量は際限なく大きくなります．
一方，一般相対論で出てくるTOV方程式を使った相対論的な計算では，中性子星の質量は最大値を持つことがわかります．
ただし，よく観測されている中性子星の質量は$1.4 M_{\odot}$程度で，最大質量は$2 M_{\odot}$程度なので，観測を説明できないことがわかります．
色々と怪しい近似をしているので完全に言い切ることはできませんが，核子間の相互作用を取り込んだ状態方程式が本質的に重要であることを示唆しています．

本題に入って，ChEFTによるハミルトニアンから出発して，IMSRG計算することで$\varepsilon(\rho)$を計算することを考えましょう．
といってもここまでの講義内容から材料はほとんど揃っていて，どのような基底関数を設定するかのみが非自明な点です．
最もシンプルな設定として，周期境界条件を課した箱の中で$\varepsilon(\rho)$を計算することを考えます．
箱の中で，一粒子状態は以下のようにラベルできます．
\begin{equation*}
|p\rangle = |\vec{k}, s_{z}, \tau_{z}\rangle, \quad \vec{k} = \frac{2\pi}{L} (n_{x}, n_{y}, n_{z}), \quad n_{i} \in \mathbb{Z}, \quad s_{z} = \pm 1/2, \quad \tau_{z} = \pm 1/2
\end{equation*}
ここで，$L$は箱の一辺の長さです．
運動量は$(n_{x}, n_{y}, n_{z})$の格子点でラベルできますが，$k_{x}^{2} + k_{y}^{2} + k_{z}^{2} \le k_{F}^{2}$となるところまでの格子点を使ってFermi球を形成します．
$k_{F}$内に存在する格子点の数はそのまま粒子数$A$になるので，密度$\rho$が与えられた時に，箱の長さ$L$は
\begin{equation*}
L = \left(\frac{A}{\rho}\right)^{1/3}
\end{equation*}
で与えられます．
このとき，ハミルトニアンは
\begin{equation*}
H = \sum_{p_{1}p_{2}} T_{p_{1}p_{2}} a^{\dag}_{p_{1}} a_{p_{2}} + \frac{1}{4} \sum_{p_{1}p_{2}p_{3}p_{4}} V_{p_{1}p_{2}p_{3}p_{4}} a^{\dag}_{p_{1}} a^{\dag}_{p_{2}} a_{p_{4}} a_{p_{3}}+ \frac{1}{36} \sum_{p_{1}p_{2}p_{3}p_{4}p_{5}p_{6}} V_{p_{1}p_{2}p_{3}p_{4}p_{5}p_{6}} a^{\dag}_{p_{1}} a^{\dag}_{p_{2}} a^{\dag}_{p_{3}} a_{p_{6}} a_{p_{5}} a_{p_{4}}
\end{equation*}
です．
ここで，$T_{p_{1}p_{2}}$，$V_{p_{1}p_{2}p_{3}p_{4}}$，$V_{p_{1}p_{2}p_{3}p_{4}p_{5}p_{6}}$はそれぞれ運動エネルギー項、二体相互作用項、三体相互作用項です．
ここで，重要な性質として，以下があります．

$T_{p_{1}p_{2}}$は運動量保存則に従うので，$\vec{k}_{1} = \vec{k}_{2}$かつ$\tau_{z,1}=\tau_{z,2}$の時にのみ非ゼロになります．

$V_{p_{1}p_{2}p_{3}p_{4}}$について，重心運動量は保存されるので，$\vec{k}_{1} + \vec{k}_{2} = \vec{k}_{3} + \vec{k}_{4}$かつ$\tau_{z,1}+\tau_{z,2}=\tau_{z,3}+\tau_{z,4}$の時にのみ非ゼロになります．

$V_{p_{1}p_{2}p_{3}p_{4}p_{5}p_{6}}$について，重心運動量は保存されるので，$\vec{k}_{1} + \vec{k}_{2} + \vec{k}_{3} = \vec{k}_{4} + \vec{k}_{5} + \vec{k}_{6}$かつ$\tau_{z,1}+\tau_{z,2}+\tau_{z,3}=\tau_{z,4}+\tau_{z,5}+\tau_{z,6}$の時にのみ非ゼロになります．

上の条件は数値計算を実装する上でとても重要な条件です．
特に，異なる重心運動量を持つ状態は混ざらないので，各重心運動量ごとのブロックに分けて計算・ストアすることで数値計算が可能になります．
いま，基準状態$|\Phi\rangle$としてFermi球を形成する状態を考えます．
$|\Phi\rangle$についてnormal orderingを定義して，2体項までを取り出すと，
\begin{equation*}
H \approx E_{0} + \sum_{p_{1}p_{2}} f_{p_{1}p_{2}} \{a^{\dag}_{p_{1}} a_{p_{2}}\} + \frac{1}{4} \sum_{p_{1}p_{2}p_{3}p_{4}} \Gamma_{p_{1}p_{2}p_{3}p_{4}} \{a^{\dag}_{p_{1}} a^{\dag}_{p_{2}} a_{p_{4}} a_{p_{3}}\}
\end{equation*}
となります．
ちなみに，今使っている基底だと$f_{p_{1}p_{2}}$について，$|\vec{k}_{1}| > k_{F}$かつ$|\vec{k}_{2}| < k_{F}$の行列要素は運動量の保存から自動的に$0$となって$|\Phi \rangle$はHartree-Fock状態に対応しています．
あとは$|\Phi\rangle$から$|\Phi^{ab}_{ij}\rangle$を分離するようにgeneratorを設定してIMSRG計算をすれば，$\varepsilon(\rho) = E_{0}(s=\infty)/A $として核物質のエネルギーを計算することができます．

実際にIMSRGを使った計算をしてみましょう．まずはソースコードの場所を見つけてコンパイルします．

```python
from pathlib import Path
import os
import shutil
import subprocess
import sys

REPO_NAME = "Lecture_Hokkaido"
REPO_URL = "https://github.com/Takayuki-Miyagi/Lecture_Hokkaido.git"

# ============================================================
# local or Colab ?
# ============================================================

try:
    import google.colab
    IN_COLAB = True
except ImportError:
    IN_COLAB = False

print("Environment:", "Google Colab" if IN_COLAB else "Local")

def find_repo_root(start: Path) -> Path:
    start = start.resolve()

    for path in [start, *start.parents]:
        if (
            (path / "pyproject.toml").exists()
            and (path / "src" / "matter" / "CMakeLists.txt").exists()
        ):
            return path

    raise FileNotFoundError(
        "Repo root not found."
    )

if IN_COLAB:
    repo_root = Path("/content") / REPO_NAME

    if not repo_root.exists():
        subprocess.run(
            ["git", "clone", REPO_URL, str(repo_root)],
            check=True,
        )
    else:
        print(f"Repository already exists: {repo_root}")

else:
    repo_root = find_repo_root(Path.cwd())

os.chdir(repo_root)

print("Repository root:", repo_root)
print("Current directory:", Path.cwd())

if IN_COLAB:
    subprocess.run(
        ["apt-get", "update", "-qq"],
        check=True,
    )

    subprocess.run(
        [
            "apt-get", "install", "-y", "-qq",
            "g++",
            "cmake",
            "ninja-build",
            "libeigen3-dev",
        ],
        check=True,
    )

else:
    required_commands = ["cmake"]

    missing_commands = [
        command
        for command in required_commands
        if shutil.which(command) is None
    ]

    if missing_commands:
        raise RuntimeError(
            "The following commands are missing: "
            + ", ".join(missing_commands)
        )

    print("Using local environment's CMake.")

source_dir = repo_root / "src" / "matter"
build_dir = repo_root / "build" / "matter"

build_dir.mkdir(parents=True, exist_ok=True)

cmake_configure_command = [
    "cmake",
    "-S", str(source_dir),
    "-B", str(build_dir),
    "-DCMAKE_BUILD_TYPE=Release",
]

if shutil.which("ninja") is not None:
    cmake_configure_command += ["-G", "Ninja"]

print("Configure command:")
print(" ".join(cmake_configure_command))

subprocess.run(
    cmake_configure_command,
    check=True,
)

subprocess.run(
    [
        "cmake",
        "--build", str(build_dir),
        "--parallel", "2",
    ],
    check=True,
)

candidate_paths = [
    build_dir / "imsrg_matter.exe",
    build_dir / "Release" / "imsrg_matter.exe",
    build_dir / "src" / "imsrg_matter.exe",
]

exe_path = next(
    (path for path in candidate_paths if path.exists()),
    None,
)

if exe_path is None:
    matches = list(build_dir.rglob("imsrg_matter.exe"))

    if matches:
        exe_path = matches[0]
    else:
        raise FileNotFoundError(
            f"imsrg_matter.exe is not found in {build_dir}"
        )

print("Executable:", exe_path)
```

これで計算する準備が整いました．

```python
params = {}
params["ch_order"] = 2
params["Ct1S0_pp"] = -0.338142
params["Ct1S0_np"] = -0.339250
params["Ct1S0_nn"] = -0.338746
params["Ct3S1"] = -0.259839
params["C1S0"] = 2.505389
params["C3S1"] = 1.002189
params["C1P1"] = -0.387960
params["C3P0"] = 0.700499
params["C3P1"] = -0.964856
params["C3SD1"] = 0.452523
params["C3P2"] = -0.883122
params["c1"] = -0.74
params["c2"] = -0.49
params["c3"] = -0.65
params["c4"] = 0.96
params["cD"] = 0.081
params["cE"] = -0.002

params["chiral_delta"] = "true"
params["include_3NF"] = "true"
params["mode"]="pnm"
params["truncation_method"] = "shell"
params["A"] = 14

params["Nmax"]=4
params["rho"]=0.16
params["RegLambda"]=394
params["reg_power"]=4
params["SFRLambda"]=700

command = [ str(exe_path)]
for key, val in params.items():
    command.append(f"{key}={val}")
subprocess.run(
    command,
    cwd=build_dir,
    text=True,
    check=True,
)
```
